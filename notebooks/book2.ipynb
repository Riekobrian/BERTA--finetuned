```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
from scipy.spatial.distance import pdist, squareform

def create_word_sentiment_network(positive_reviews, negative_reviews):
    # Combine all reviews
    all_reviews = positive_reviews + ' ' + negative_reviews
    
    # Create a CountVectorizer to extract words
    vectorizer = CountVectorizer(stop_words='english', max_features=100)
    word_matrix = vectorizer.fit_transform([all_reviews])
    
    # Get the words
    words = vectorizer.get_feature_names_out()
    
    # Create a graph
    G = nx.Graph()
    
    # Color mapping for sentiments
    def classify_word_sentiment(word):
        # Simple sentiment classification (you'd typically use a more sophisticated method)
        positive_words = set(['good', 'great', 'excellent', 'amazing', 'wonderful', 'love', 'best'])
        negative_words = set(['bad', 'worst', 'terrible', 'awful', 'horrible', 'poor', 'hate'])
        
        if word in positive_words:
            return 'green'
        elif word in negative_words:
            return 'red'
        else:
            return 'yellow'
    
    # Add nodes with sentiment-based colors
    for word in words:
        color = classify_word_sentiment(word)
        G.add_node(word, color=color)
    
    # Create connections based on co-occurrence
    word_freq = word_matrix.toarray()[0]
    
    # Create a co-occurrence matrix
    co_occurrence = np.zeros((len(words), len(words)))
    for i in range(len(words)):
        for j in range(i+1, len(words)):
            # Simple co-occurrence calculation
            co_occurrence[i,j] = min(word_freq[i], word_freq[j])
            co_occurrence[j,i] = co_occurrence[i,j]
    
    # Add edges based on co-occurrence
    for i in range(len(words)):
        for j in range(i+1, len(words)):
            if co_occurrence[i,j] > 0:
                G.add_edge(words[i], words[j], weight=co_occurrence[i,j])
    
    # 3D Visualization
    plt.figure(figsize=(20,15))
    
    # Use spring layout for node positioning
    pos = nx.spring_layout(G, k=0.5, iterations=50, seed=42)
    
    # Draw nodes
    node_colors = [G.nodes[node]['color'] for node in G.nodes()]
    node_sizes = [G.degree(node) * 100 for node in G.nodes()]
    
    # Draw the network
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes, alpha=0.7)
    nx.draw_networkx_edges(G, pos, alpha=0.2, width=[G[u][v]['weight']/10 for u,v in G.edges()])
    
    # Add labels
    nx.draw_networkx_labels(G, pos, font_size=8, font_weight="bold")
    
    plt.title("3D Word Sentiment Network", fontsize=20)
    plt.axis('off')
    
    # Additional network analysis
    print("Network Statistics:")
    print(f"Total Words: {G.number_of_nodes()}")
    print(f"Total Connections: {G.number_of_edges()}")
    
    # Sentiment distribution
    sentiment_colors = {'green': 'Positive', 'red': 'Negative', 'yellow': 'Neutral'}
    color_counts = {}
    for color in sentiment_colors:
        color_counts[sentiment_colors[color]] = node_colors.count(color)
    
    print("\nWord Sentiment Distribution:")
    for sentiment, count in color_counts.items():
        print(f"{sentiment} Words: {count}")
    
    plt.tight_layout()
    plt.show()
    
    return G, pos

# Example usage (you'd replace this with your actual dataframe)
positive_reviews = "This product is amazing and works great. I love how efficient it is and the quality is excellent."
negative_reviews = "The product is terrible and doesn't work at all. I hate how poor the design is and it's the worst purchase ever."

# Create the network
network, positions = create_word_sentiment_network(positive_reviews, negative_reviews)
```

TFIDF = TfidfVectorizer

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer


def clean_text_for_vectorization(X_train, X_test=None):
    """
    Comprehensive text cleaning function

    Parameters:
    - X_train: Training text data
    - X_test: Optional test text data

    Returns:
    - Cleaned training data
    - Cleaned test data (if provided)
    """
    # Replace NaN with empty string
    X_train_cleaned = X_train.fillna("").astype(str)

    # Additional cleaning steps
    X_train_cleaned = X_train_cleaned.str.lower()  # Convert to lowercase
    X_train_cleaned = X_train_cleaned.str.replace(
        r"[^a-zA-Z\s]", "", regex=True
    )  # Remove special characters

    if X_test is not None:
        X_test_cleaned = X_test.fillna("").astype(str)
        X_test_cleaned = X_test_cleaned.str.lower()
        X_test_cleaned = X_test_cleaned.str.replace(r"[^a-zA-Z\s]", "", regex=True)
        return X_train_cleaned, X_test_cleaned

    return X_train_cleaned


def advanced_tfidf_vectorization(
    X_train, X_test, max_features=5000, ngram_range=(1, 2), stop_words="english"
):
    """
    Advanced TF-IDF Vectorization with comprehensive error handling and analysis

    Parameters:
    - X_train: Training text data
    - X_test: Test text data
    - max_features: Maximum number of features
    - ngram_range: Range of n-grams to consider
    - stop_words: Stop words removal strategy

    Returns:
    - Vectorized training data
    - Vectorized test data
    - Vectorizer object
    """
    # Clean the data
    X_train_cleaned, X_test_cleaned = clean_text_for_vectorization(X_train, X_test)

    # Create TF-IDF Vectorizer
    tfidf_vectorizer = TfidfVectorizer(
        max_features=max_features, ngram_range=ngram_range, stop_words=stop_words
    )

    # Fit and transform
    X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_cleaned)
    X_test_tfidf = tfidf_vectorizer.transform(X_test_cleaned)

    # Detailed feature analysis
    feature_names = tfidf_vectorizer.get_feature_names_out()
    print(f"\nTotal Features: {len(feature_names)}")

    # Get top features by TF-IDF score
    feature_importances = np.asarray(X_train_tfidf.sum(axis=0)).ravel()
    top_features = sorted(
        zip(feature_names, feature_importances), key=lambda x: x[1], reverse=True
    )[:20]

    print("\nTop 20 Most Important Features:")
    for feature, importance in top_features:
        print(f"{feature}: {importance:.4f}")

    return X_train_tfidf, X_test_tfidf, tfidf_vectorizer


def diagnose_text_data(X_train, X_test):
    """
    Comprehensive diagnostic function for text data

    Parameters:
    - X_train: Training text data
    - X_test: Test text data

    Prints detailed information about the dataset
    """
    print("\n--- Data Quality Diagnostics ---")

    # NaN Analysis
    print("\nNaN Value Analysis:")
    print(f"NaN values in X_train: {X_train.isna().sum()}")
    print(f"NaN values in X_test: {X_test.isna().sum()}")

    # Data Type and Sample Analysis
    print("\nData Type Analysis:")
    print(f"X_train data type: {X_train.dtype}")
    print(f"X_test data type: {X_test.dtype}")

    # Sample of unique values
    print("\nSample Unique Values (X_train):")
    print(X_train.dropna().unique()[:5])

    # Text Length Analysis
    print("\nText Length Analysis:")
    print("X_train text lengths:")
    print(X_train.str.len().describe())
    print("\nX_test text lengths:")
    print(X_test.str.len().describe())


def tfidf_vectorization_workflow(X_train, X_test):
    """
    Complete workflow for TF-IDF vectorization

    Parameters:
    - X_train: Training text data
    - X_test: Test text data

    Returns:
    - Vectorized training data
    - Vectorized test data
    - Vectorizer object
    """
    # Diagnostic check
    diagnose_text_data(X_train, X_test)

    # Vectorization
    X_train_tfidf, X_test_tfidf, vectorizer = advanced_tfidf_vectorization(
        X_train, X_test
    )

    return X_train_tfidf, X_test_tfidf, vectorizer


# Example Usage (replace with your actual data)
# X_train_tfidf, X_test_tfidf, vectorizer = tfidf_vectorization_workflow(X_train, X_test)
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


def clean_text_data(data):
    """
    Comprehensive text cleaning function

    Parameters:
    - data: Input text data (Series or list)

    Returns:
    - Cleaned text data as list of strings
    """
    # Convert to string and handle NaN values
    cleaned_data = data.fillna("").astype(str)

    # Optional: Additional cleaning steps
    def text_cleaner(text):
        # Convert to lowercase
        text = text.lower()

        # Remove special characters (optional)
        # text = re.sub(r'[^a-zA-Z\s]', '', text)

        return text

    # Apply cleaning to each text
    cleaned_data = cleaned_data.apply(text_cleaner)

    return cleaned_data.tolist()


def advanced_text_tokenization(
    X_train, X_test, num_words=5000, max_length=100, oov_token="<OOV>"
):
    """
    Advanced text tokenization with robust error handling

    Parameters:
    - X_train: Training text data
    - X_test: Test text data
    - num_words: Maximum number of words to keep
    - max_length: Maximum sequence length
    - oov_token: Out of vocabulary token

    Returns:
    - Padded training sequences
    - Padded test sequences
    - Tokenizer object
    """
    # Clean the text data
    X_train_cleaned = clean_text_data(X_train)
    X_test_cleaned = clean_text_data(X_test)

    # Create tokenizer
    tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)

    # Fit tokenizer on training data
    tokenizer.fit_on_texts(X_train_cleaned)

    # Convert texts to sequences
    X_train_seq = tokenizer.texts_to_sequences(X_train_cleaned)
    X_test_seq = tokenizer.texts_to_sequences(X_test_cleaned)

    # Pad sequences
    X_train_pad = pad_sequences(
        X_train_seq, maxlen=max_length, padding="post", truncating="post"
    )
    X_test_pad = pad_sequences(
        X_test_seq, maxlen=max_length, padding="post", truncating="post"
    )

    # Print some diagnostic information
    print("Tokenization Diagnostics:")
    print(f"Total words in vocabulary: {len(tokenizer.word_index)}")
    print(f"Training sequences shape: {X_train_pad.shape}")
    print(f"Test sequences shape: {X_test_pad.shape}")

    # Show top 20 most frequent words
    word_counts = tokenizer.word_counts
    top_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)[:20]

    print("\nTop 20 Most Frequent Words:")
    for word, count in top_words:
        print(f"{word}: {count}")

    return X_train_pad, X_test_pad, tokenizer


# Diagnostic function to check data quality
def check_text_data_quality(X_train, X_test):
    """
    Perform comprehensive text data quality checks
    """
    print("\nText Data Quality Diagnostics:")

    # Check for NaN values
    print(f"NaN values in X_train: {X_train.isna().sum()}")
    print(f"NaN values in X_test: {X_test.isna().sum()}")

    # Check data types
    print(f"\nData types in X_train: {X_train.dtype}")
    print(f"Data types in X_test: {X_test.dtype}")

    # Sample of first few values
    print("\nSample of first few values in X_train:")
    print(X_train.dropna().head())


# Recommended workflow function
def tokenization_workflow(X_train, X_test):
    """
    Workflow for text tokenization and preprocessing

    Args:
      X_train: Training text data.
      X_test: Test text data.

    Returns:
      X_train_pad: Padded training sequences.
      X_test_pad: Padded test sequences.
      tokenizer: Fitted Tokenizer object.
    """
    # 1. Check data quality
    check_text_data_quality(X_train, X_test)

    # 2. Clean the text data
    X_train_cleaned = clean_text_data(X_train)
    X_test_cleaned = clean_text_data(X_test)

    # 3. Create and fit the tokenizer
    tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
    tokenizer.fit_on_texts(X_train_cleaned)

    # 4. Convert text to sequences
    X_train_seq = tokenizer.texts_to_sequences(X_train_cleaned)
    X_test_seq = tokenizer.texts_to_sequences(X_test_cleaned)

    # 5. Pad the sequences
    X_train_pad = pad_sequences(
        X_train_seq, maxlen=100, padding="post", truncating="post"
    )
    X_test_pad = pad_sequences(
        X_test_seq, maxlen=100, padding="post", truncating="post"
    )

    # 6. Print diagnostic information
    print("Tokenization Diagnostics:")
    print(f"Total words in vocabulary: {len(tokenizer.word_index)}")
    print(f"Training sequences shape: {X_train_pad.shape}")
    print(f"Test sequences shape: {X_test_pad.shape}")

    # 7. Show top 20 most frequent words
    word_counts = tokenizer.word_counts
    top_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)[:20]
    print("\nTop 20 Most Frequent Words:")
    for word, count in top_words:
        print(f"{word}: {count}")

    return X_train_pad, X_test_pad, tokenizer


# Example usage (you would replace this with your actual data)
# X_train_pad, X_test_pad, tokenizer = tokenization_workflow(X_train, X_test)
```

**#Define and Train LSTM Model**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

def prepare_lstm_model(X_train_pad, num_features=5000):
    # Determine max_length dynamically
    max_length = X_train_pad.shape[1]
    
    # Model Architecture
    model = Sequential([
        # Embedding layer
        Embedding(input_dim=num_features, 
                  output_dim=128, 
                  input_length=max_length,
                  mask_zero=True),  # Handle variable-length sequences
        
        # First LSTM layer with return_sequences to feed into next LSTM
        LSTM(64, 
             return_sequences=True, 
             dropout=0.3,  # Dropout within LSTM layer
             recurrent_dropout=0.3),  # Dropout for recurrent connections
        
        # Dropout for regularization
        Dropout(0.5),
        
        # Second LSTM layer
        LSTM(32, 
             dropout=0.3, 
             recurrent_dropout=0.3),
        
        # Dense layers for classification
        Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        Dropout(0.4),
        
        # Output layer for binary classification
        Dense(1, activation='sigmoid')
    ])
    
    # Compile the model with advanced optimization
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy', 
        metrics=['accuracy', 
                 tf.keras.metrics.Precision(), 
                 tf.keras.metrics.Recall()]
    )
    
    return model

def train_model_with_callbacks(model, X_train_pad, y_train, X_test_pad, y_test):
    # Callbacks for improved training
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=5,  # Number of epochs with no improvement
        restore_best_weights=True,
        verbose=1
    )
    
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,  # Reduce learning rate by 80%
        patience=3,
        min_lr=0.00001,
        verbose=1
    )
    
    # Model Checkpoint callback
    checkpoint = ModelCheckpoint(
        filepath='checkpoint_model.h5',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
    
    # Train the model with callbacks
    history = model.fit(
        X_train_pad, y_train, 
        epochs=20,  # Increased epochs with early stopping
        batch_size=32, 
        validation_data=(X_test_pad, y_test),
        callbacks=[early_stopping, reduce_lr, checkpoint]
    )
    
    return history

def evaluate_model(model, X_test_pad, y_test):
    """
    Comprehensive model evaluation
    """
    # Evaluate the model
    test_loss, test_accuracy, test_precision, test_recall = model.evaluate(X_test_pad, y_test, verbose=0)
    
    print("\nModel Evaluation:")
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Test Precision: {test_precision:.4f}")
    print(f"Test Recall: {test_recall:.4f}")
    
    # Predictions
    y_pred = model.predict(X_test_pad)
    y_pred_binary = (y_pred > 0.5).astype(int)
    
    # Confusion Matrix
    from sklearn.metrics import confusion_matrix, classification_report
    cm = confusion_matrix(y_test, y_pred_binary)
    print("\nConfusion Matrix:")
    print(cm)
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_binary))
    
    return test_loss, test_accuracy

# Workflow function
def lstm_classification_workflow(X_train_pad, y_train, X_test_pad, y_test):
    """
    Complete workflow for LSTM text classification
    """
    # Create model
    model = prepare_lstm_model(X_train_pad)
    
    # Print model summary
    model.summary()
    
    # Train model
    history = train_model_with_callbacks(model, X_train_pad, y_train, X_test_pad, y_test)
    
    # Save the trained model
    model.save('final_lstm_model.h5')
    print("Model saved as 'final_lstm_model.h5'")
    
    # Evaluate model
    evaluate_model(model, X_test_pad, y_test)
    
    return model, history

# Example usage (you would replace with your actual padded data)
# model, history = lstm_classification_workflow(X_train_pad, y_train, X_test_pad, y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Bidirectional, Dense, Embedding, GlobalMaxPooling1D
from sklearn.model_selection import ParameterGrid
import tensorflow as tf

def create_model(architecture='lstm', 
                 embedding_dim=128, 
                 lstm_units=64, 
                 dropout_rate=0.3, 
                 vocab_size=10000, 
                 max_length=100):
    """
    Create different neural network architectures for comparison
    
    Parameters:
    - architecture: 'lstm', 'gru', or 'bilstm'
    - embedding_dim: Embedding layer dimension
    - lstm_units: Number of units in LSTM/GRU layer
    - dropout_rate: Dropout rate for regularization
    - vocab_size: Size of vocabulary
    - max_length: Maximum sequence length
    """
    model = Sequential()
    model.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
    
    if architecture == 'lstm':
        model.add(LSTM(lstm_units, return_sequences=False, dropout=dropout_rate))
    elif architecture == 'gru':
        model.add(GRU(lstm_units, return_sequences=False, dropout=dropout_rate))
    elif architecture == 'bilstm':
        model.add(Bidirectional(LSTM(lstm_units, return_sequences=False, dropout=dropout_rate)))
    
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(optimizer='adam', 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    
    return model

# Hyperparameter Grid
param_grid = {
    'architecture': ['lstm', 'gru', 'bilstm'],
    'embedding_dim': [64, 128, 256],
    'lstm_units': [64, 128, 256],
    'dropout_rate': [0.2, 0.3, 0.4]
}

# Comparative Analysis Function
def comparative_analysis(X_train, y_train, X_val, y_val):
    results = []
    
    for params in ParameterGrid(param_grid):
        model = create_model(
            architecture=params['architecture'],
            embedding_dim=params['embedding_dim'],
            lstm_units=params['lstm_units'],
            dropout_rate=params['dropout_rate']
        )
        
        history = model.fit(
            X_train, y_train, 
            validation_data=(X_val, y_val),
            epochs=10, 
            batch_size=32,
            verbose=0
        )
        
        val_loss, val_accuracy = model.evaluate(X_val, y_val)
        
        results.append({
            'params': params,
            'val_loss': val_loss,
            'val_accuracy': val_accuracy
        })
    
    return sorted(results, key=lambda x: x['val_accuracy'], reverse=True)

```

```python
import flask
from flask import Flask, request, jsonify
from threading import Thread
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
import pickle  # Ensure pickle is imported

app = Flask(__name__)

# Global variables for preprocessing
max_length = 100  # Match the max length used during training
vocab_size = 10000  # Match the vocab size used during training

# Load tokenizer (if you saved it)
try:
    with open('tokenizer.pickle', 'rb') as handle:
        tokenizer = pickle.load(handle)
except FileNotFoundError:
    # Fallback: Create a new tokenizer if not saved
    tokenizer = Tokenizer(num_words=vocab_size)

# Load your trained model
try:
    model = tf.keras.models.load_model('final_lstm_model.h5')
except Exception as e:
    print(f"Error loading model: {e}")
    model = None

def preprocess_text(text):
    """
    Preprocess input text for model prediction
    
    Args:
        text (str): Input text to preprocess
    
    Returns:
        numpy.ndarray: Preprocessed and padded sequence
    """
    if not text:
        raise ValueError("Input text cannot be empty")
    
    # Tokenize the text
    sequences = tokenizer.texts_to_sequences([text])
    
    # Pad sequences to match model's expected input
    padded_sequence = pad_sequences(sequences, maxlen=max_length)
    
    return padded_sequence

@app.route('/')
def home():
    return render_template('index.html')
    """
    Endpoint for text prediction
    
    Returns:
        JSON response with prediction details
    """
    if not model:
        return jsonify({
            'error': 'Model not loaded correctly',
            'status': 'failure'
        }), 500
    
    try:
        data = request.json
        if not data:
            return jsonify({
                'error': 'No input data provided',
                'status': 'failure'
            }), 400
        
        text = data.get('text', '')
        if not text:
            return jsonify({
                'error': 'Text field is empty',
                'status': 'failure'
            }), 400
        
        # Preprocess input
        processed_input = preprocess_text(text)
        
        # Make prediction
        prediction = model.predict(processed_input)
        
        # Convert prediction to human-readable format
        result = {
            'text': text,
            'prediction': float(prediction[0][0]),
            'class': 'Positive' if prediction[0][0] > 0.5 else 'Negative',
            'status': 'success'
        }
        
        return jsonify(result)
    
    except Exception as e:
        app.logger.error(f"Prediction error: {str(e)}")
        return jsonify({
            'error': str(e),
            'status': 'failure'
        }), 500

@app.route('/health', methods=['GET'])
def health_check():
    """
    Health check endpoint to verify service is running
    """
    return jsonify({
        'status': 'healthy',
        'model_loaded': model is not None
    })

# Run Flask in a separate thread
def run_flask():
    app.run(host='0.0.0.0', port=5000, debug=False)

flask_thread = Thread(target=run_flask)
flask_thread.daemon = True  # Ensures thread closes when the notebook stops
flask_thread.start()

print("Flask app is running...")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    confusion_matrix, 
    classification_report, 
    precision_recall_curve, 
    roc_curve, 
    auc
)

def visualize_model_performance(model, X_test, y_test):
    # Predictions
    y_pred = model.predict(X_test)
    y_pred_binary = (y_pred > 0.5).astype(int)
    
    # 1. Confusion Matrix
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    cm = confusion_matrix(y_test, y_pred_binary)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    
    # 2. ROC Curve
    plt.subplot(1, 2, 2)
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, color='darkorange', lw=2, 
             label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    
    plt.tight_layout()
    plt.show()
    
    # 3. Detailed Classification Report
    print(classification_report(y_test, y_pred_binary))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    classification_report, 
    confusion_matrix, 
    accuracy_score, 
    precision_score, 
    recall_score, 
    f1_score
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Embedding, 
    LSTM, 
    Dense, 
    Dropout, 
    GlobalMaxPooling1D
)
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

class SentimentAnalysisWorkflow:
    def __init__(self, 
                 max_words=5000, 
                 max_len=200, 
                 embedding_dim=100):
        """
        Initialize sentiment analysis workflow
        
        Parameters:
        - max_words: Maximum number of words to keep
        - max_len: Maximum sequence length
        - embedding_dim: Dimensionality of word embeddings
        """
        self.max_words = max_words
        self.max_len = max_len
        self.embedding_dim = embedding_dim
        
        # Initialize tokenizer and label encoder
        self.tokenizer = Tokenizer(num_words=max_words, oov_token='<OOV>')
        self.label_encoder = LabelEncoder()
    
    def preprocess_text(self, text_series):
        """
        Clean and preprocess text data
        
        Parameters:
        - text_series: Pandas Series of text data
        
        Returns:
        - Cleaned text series
        """
        # Clean text: remove NaN, convert to string
        cleaned_text = text_series.fillna('').astype(str)
        
        # Optional: Additional text cleaning steps
        cleaned_text = cleaned_text.str.lower()
        
        return cleaned_text
    
    def tokenize_texts(self, X_train, X_test):
        """
        Tokenize texts and create padded sequences
        
        Parameters:
        - X_train: Training text data
        - X_test: Test text data
        
        Returns:
        - Padded training sequences
        - Padded test sequences
        """
        # Preprocess text
        X_train_cleaned = self.preprocess_text(X_train)
        X_test_cleaned = self.preprocess_text(X_test)
        
        # Fit tokenizer on training data
        self.tokenizer.fit_on_texts(X_train_cleaned)
        
        # Convert texts to sequences
        X_train_seq = self.tokenizer.texts_to_sequences(X_train_cleaned)
        X_test_seq = self.tokenizer.texts_to_sequences(X_test_cleaned)
        
        # Pad sequences
        X_train_pad = pad_sequences(
            X_train_seq, 
            maxlen=self.max_len, 
            padding='post', 
            truncating='post'
        )
        X_test_pad = pad_sequences(
            X_test_seq, 
            maxlen=self.max_len, 
            padding='post', 
            truncating='post'
        )
        
        return X_train_pad, X_test_pad
    
    def encode_labels(self, y_train, y_test):
        """
        Encode categorical labels
        
        Parameters:
        - y_train: Training labels
        - y_test: Test labels
        
        Returns:
        - Encoded training labels
        - Encoded test labels
        """
        # Fit label encoder on training data
        y_train_encoded = self.label_encoder.fit_transform(y_train)
        y_test_encoded = self.label_encoder.transform(y_test)
        
        # Convert to categorical
        y_train_cat = tf.keras.utils.to_categorical(y_train_encoded)
        y_test_cat = tf.keras.utils.to_categorical(y_test_encoded)
        
        return y_train_cat, y_test_cat
    
    def create_lstm_model(self, input_shape, num_classes):
        """
        Create LSTM model for sentiment analysis
        
        Parameters:
        - input_shape: Shape of input sequences
        - num_classes: Number of sentiment classes
        
        Returns:
        - Compiled Keras model
        """
        model = Sequential([
            Embedding(
                input_dim=self.max_words, 
                output_dim=self.embedding_dim, 
                input_length=input_shape[1]
            ),
            LSTM(128, return_sequences=True),
            GlobalMaxPooling1D(),
            Dense(64, activation='relu'),
            Dropout(0.5),
            Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam', 
            loss='categorical_crossentropy', 
            metrics=['accuracy']
        )
        
        return model
    
    def train_model(self, X_train, y_train, X_val, y_val):
        """
        Train the sentiment analysis model
        
        Parameters:
        - X_train: Training input sequences
        - y_train: Training labels
        - X_val: Validation input sequences
        - y_val: Validation labels
        
        Returns:
        - Trained model
        - Training history
        """
        # Create model
        model = self.create_lstm_model(
            input_shape=X_train.shape, 
            num_classes=y_train.shape[1]
        )
        
        # Define callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss', 
            patience=5, 
            restore_best_weights=True
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss', 
            factor=0.2,
            patience=3, 
            min_lr=0.0001
        )
        
        # Train model
        history = model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=50,
            batch_size=64,
            callbacks=[early_stopping, reduce_lr],
            verbose=1
        )
        
        return model, history
    
    def evaluate_model(self, model, X_test, y_test):
        """
        Comprehensive model evaluation
        
        Parameters:
        - model: Trained Keras model
        - X_test: Test input sequences
        - y_test: Test labels
        """
        # Predict on test data
        y_pred_prob = model.predict(X_test)
        y_pred = np.argmax(y_pred_prob, axis=1)
        y_true = np.argmax(y_test, axis=1)
        
        # Generate comprehensive report
        print("\nModel Evaluation Report:")
        print(classification_report(
            y_true, 
            y_pred, 
            target_names=self.label_encoder.classes_
        ))
        
        # Confusion Matrix
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(
            cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=self.label_encoder.classes_,
            yticklabels=self.label_encoder.classes_
        )
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.tight_layout()
        plt.savefig('confusion_matrix.png')
        plt.close()
    
    def plot_training_history(self, history):
        """
        Plot training history
        
        Parameters:
        - history: Keras training history
        """
        plt.figure(figsize=(12, 4))
        
        # Accuracy subplot
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        
        # Loss subplot
        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        
        plt.tight_layout()
        plt.savefig('training_history.png')
        plt.close()
    
    def run_workflow(self, X, y, test_size=0.2, val_size=0.2):
        """
        Complete workflow for sentiment analysis
        
        Parameters:
        - X: Text features
        - y: Labels
        - test_size: Proportion of data for testing
        - val_size: Proportion of training data for validation
        """
        # Initial split
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, 
            test_size=test_size, 
            random_state=42
        )
        
        # Further split training data for validation
        X_train, X_val, y_train, y_val = train_test_split(
            X_train, y_train, 
            test_size=val_size, 
            random_state=42
        )
        
        # Tokenize texts
        X_train_pad, X_test_pad = self.tokenize_texts(X_train, X_test)
        X_val_pad, _ = self.tokenize_texts(X_val, X_test)
        
        # Encode labels
        y_train_cat, y_test_cat = self.encode_labels(y_train, y_test)
        y_val_cat, _ = self.encode_labels(y_val, y_test)
        
        # Train model
        model, history = self.train_model(
            X_train_pad, y_train_cat, 
            X_val_pad, y_val_cat
        )
        
        # Evaluate model
        self.evaluate_model(model, X_test_pad, y_test_cat)
        
        # Plot training history
        self.plot_training_history(history)
        
        return model

```
